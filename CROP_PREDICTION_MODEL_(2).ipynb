```python
#split features & labels
x = df.iloc[:,:-1] #features
x
```

```python
y = df.iloc[:,-1] #labels
y
```

```python
#We have splitted the data into training and testing
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
x_train.head() #Display the training data of only the features
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 5))
sns.countplot(x='label', data=df, order=df['label'].value_counts().index, palette='Set3')
plt.xticks(rotation=90)
plt.title('Crop Frequency Distribution')
plt.xlabel('Crop')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 10))
df['label'].value_counts().plot.pie(autopct='%1.1f%%', startangle=140, cmap='tab20')
plt.title('Crop Distribution (Pie Chart)')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.boxplot(data=df, x='label', y='ph', palette='coolwarm')
plt.title('pH Distribution per Crop')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
avg_rainfall = df.groupby('label')['rainfall'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_rainfall.index, y=avg_rainfall.values, palette='Blues_d')
plt.xticks(rotation=90)
plt.title('Average Rainfall per Crop')
plt.xlabel('Crop')
plt.ylabel('Rainfall (mm)')
plt.tight_layout()
plt.show()
```

```python
# Choose example crops
example_crops = ['rice', 'maize', 'cotton']
plt.figure(figsize=(12, 6))

for crop in example_crops:
    temp_series = df[df['label'] == crop]['temperature'].reset_index(drop=True)
    plt.plot(temp_series, label=crop)

plt.title('Temperature Trend per Sample (for Selected Crops)')
plt.xlabel('Sample Index')
plt.ylabel('Temperature (Â°C)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.drop('label', axis=1).corr(), annot=True, cmap='YlGnBu', linewidths=0.5)
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
model = RandomForestClassifier() #To call the ML model
model.fit(x_train,y_train) #To train the model
```

```python
predictions = model.predict(x_test) #To predict the output
```

```python
y_pred = model.predict(x_test)

from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print("Accuracy = ",accuracy_score(y_test,y_pred))
```

```python
#Testing with a new set of features i.e what will be the output if a new set of features is tested ?
new_data = [36,58,25,28.66024,59.31891,8.399136,36.92]
predicted_crop = model.predict([new_data])
predicted_crop

#Here in the above code we create a variable where we stored the new data in 2d array and called the fn
```
